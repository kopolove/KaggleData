```
import seaborn as sns
```

```
typeavg = dfgen1[["Type 1", "HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]].groupby(by=["Type 1"]).mean()
```

```
from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="white")


# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(typeavg, linewidths=.5, cmap="YlGnBu")
```

```
typeavg
```

```
typeavg = typeavg.reset_index()
```

```
typeavg
```

```
type_avg_melted = pd.melt(typeavg, id_vars=["Type 1"], var_name="Stat")
```

```
type_avg_melted
```

```
plt.figure(figsize=(12,10))
plt.ylim(0, 275)
sns.swarmplot(x="Stat", y="value", data=type_avg_melted, hue="Type 1", split=True, size=7)
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.);
```
